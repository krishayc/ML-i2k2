```python
import pandas as pd

df = pd.read_csv("sample_sales_data.csv")
print(df.head())
print(df.info())
print(df.columns)
```

```python
# 2. Fix column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# 3. Handle missing values
print(df.isnull().sum())
df["sales"] = df["sales"].fillna(df["sales"].mean())
df["quantity"] = df["quantity"].fillna(df["quantity"].mean())
df["customer_name"] = df["customer_name"].fillna(df["customer_name"].mean())
df = df.dropna(subset=["order_id"])   

# 4. Remove duplicates
df = df.drop_duplicates()
df = df.reset_index(drop=True)

# 5. Fix data types
df["order_date"] = pd.to_datetime(df["order_date"])
df["sales"] = df["sales"].astype(float)
df["quantity"] = df["quantity"].astype(float)

# 6. Clean text columns
df["category"] = df["category"].str.strip().str.title()
df["region"] = df["region"].str.strip().str.title()
df["month"] = df["month"].str.strip().str.title()
df["customer_name"] = df["customer_name"].str.strip().str.title()
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
print(df.describe())

# 9. Save
df.to_csv("sample_sales_data_cleaned_.csv", index=False)
print("Cleaning complete. Saved to sample_sales_data_cleaned_.csv")
```
