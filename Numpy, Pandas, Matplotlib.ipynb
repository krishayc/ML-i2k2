```python
import pandas as pd
```

```python
#Load a CSV file and print the number of missing values in each column.
df = pd.read_csv("sample_sales_data.csv")
print(df.isnull().sum())
```

```python
#Remove all duplicate rows from a DataFrame.
df = df.drop_duplicates()
```

```python
#Convert a column called "order_date" from text to a proper datetime type.
df["order_date"] = pd.to_datetime(df["order_date"])
print(df["order_date"].dtype) 
```

```python
#Group a sales DataFrame by "region" and find the total and average sales per region.
summary = df.groupby("region").agg(
    total_sales=("sales", "sum"),
    avg_sales=("sales", "mean")
)
print(summary)
```

```python
#Create a pivot table showing total sales by "category" (rows) and "month" (columns).
pivot = df.pivot_table(
    values="sales",
    index="category",
    columns="month",
    aggfunc="sum",
    fill_value=0
)
```

```python
#Line chart of sales over 6 months
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [1200, 1500, 1100, 1800, 2200, 2000]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker="o", color="blue")
plt.title("Sales Over 6 Months")
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.show()
```

```python
#Bar chart comparing 3 categories
categories = ["Electronics", "Clothing", "Food"]
totals = [4500, 3200, 1800]

plt.figure(figsize=(8, 5))
plt.bar(categories, totals, color="orange")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales ($)")
plt.show()
```

```python
#Pie chart of 4 product categories
labels = ["Electronics", "Clothing", "Food", "Toys"]
values = [4500, 3200, 1800, 900]

plt.figure(figsize=(6, 6))
plt.pie(values, labels=labels, autopct="%1.1f%%")
plt.title("Category Share")
plt.show()
```

```python
#Histogram of 20 random sales values
import numpy as np

random_sales = np.random.randint(500, 3000, size=20)

plt.figure(figsize=(8, 5))
plt.hist(random_sales, bins=5, color="green", edgecolor="black")
plt.title("Distribution of Sales Values")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.title("Distribution of Sales Values")
plt.show()
```
