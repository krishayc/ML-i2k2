```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Simulation of a coin toss
head = 0
tail = 0
toss = np.random.randint(0, 2, 10000)
for i in range(10000):
    if toss[i] == 0:
        head += 1
    else:
        tail += 1
       
print ("Heads=",head)
print("Tails=",tail)
print("P(Heads)=",head/10000)
print("P(Tails)=",tail/10000)
```

```python
# Simulation of a dice roll

a = [0]*6
roll = np.random.randint(1, 7, 10000)
for i in range(6):
    for j in range(10000):
        if i+1 == roll[j]:
            a[i] += 1
for i in range(6):
    print(i+1,"=",a[i])
q = int(input("Enter the number for which you want the probability: "))
print(a[q-1]/10000)
```

```python
# Conditional probability using deck of cards

np.random.seed(0)
n = 100000

ranks = np.random.choice(list(range(1,14)), size=n)   # 1-13, where 11=J,12=Q,13=K
is_face_card = ranks >= 11
is_king = ranks == 13

p_king_given_face = np.mean(is_king[is_face_card])
print(p_king_given_face)  
```

```python
from scipy.stats import binom

n, p, k = 10, 0.5, 3
prob = binom.pmf(k, n, p)
print(f"P(exactly 3 heads in 10 flips) = {prob:.4f}")

np.random.seed(5)
sims = np.random.binomial(n, p, size=100000)
print(f"Simulated P(X=3) = {np.mean(sims == 3):.4f}")

k_values = np.arange(0, 11)
probs = binom.pmf(k_values, n, p)

plt.bar(k_values, probs, edgecolor="black")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.show()
```
