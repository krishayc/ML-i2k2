```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
sample_data1 = [88, 92, 85, 90, 87, 91, 89, 84, 93, 86] 
sample_data2 = [80, 83, 78, 85, 81, 79, 84, 77, 82, 80] 
alpha = 0.05
```

```python
x, y = stats.ttest_ind(sample_data1, sample_data2)
print("t stat =", x)
print("p value =", y)

if y < alpha:
    print("Reject Null Hypothesis")
    print("The two groups are significantly different")
else:
    print("Fail to Reject Null Hypothesis")
    print("No significant difference between groups")
```

```python
std = np.std(sample_data1)
mean = np.mean(sample_data1)
n = len(sample_data1)
t_critical = stats.t.ppf((1 + (1-alpha)) / 2, df=n-1)
print(t_critical)
```

```python
margin = t_critical * (std / np.sqrt(n))
print(margin)
```

```python
print("95% Confidence Interval:")
print(mean - margin, mean + margin)
```

```python
std = np.std(sample_data2)
mean = np.mean(sample_data2)
n = len(sample_data2)
t_critical = stats.t.ppf((1 + (1-alpha)) / 2, df=n-1)
print(t_critical)
```

```python
margin = t_critical * (std / np.sqrt(n))
print(margin)
```

```python
print("95% Confidence Interval:")
print(mean - margin, mean + margin)
```
